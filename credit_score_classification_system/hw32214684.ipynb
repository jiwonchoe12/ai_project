# 데이터 전처리1

```python
df.isna().sum()
```

```python
dropdf = df.dropna(axis=0) #nan 제거
```

```python
dropdf.isna().sum()
```

```python
def prtUniq(df):
  for column in df.columns:
    print("I am", column)
    print(df[column].unique())
    print("==========================================")
```

```python
dropdf.info()
```

```python
prtUniq(dropdf)
```

```python
#이상한 값 제거
dropdf = dropdf[dropdf['Occupation'].str.contains('_______') == False]
dropdf = dropdf[dropdf['Payment_Behaviour'].str.contains('!@9#%8') == False]
dropdf = dropdf[dropdf['Credit_Mix'].str.contains('_') == False]
dropdf = dropdf[dropdf['SSN'] != '#F%$D@*&8']
```

```python
sym = "\\`*_{}[]()>#@+!$:;"
col_int = ['Age','Delay_from_due_date','Num_of_Delayed_Payment','Outstanding_Debt',
           'Total_EMI_per_month','Monthly_Balance','Annual_Income']
col_str = ['Occupation','Credit_History_Age','Payment_of_Min_Amount','Credit_Mix']
for i in col_int:
  for c in sym:
    dropdf[i] = dropdf[i].astype(str).str.replace(c,'')
for i in col_str:
  for c in sym:
    dropdf[i] = dropdf[i].replace(c,'')
dropdf.head()
```

```python
dropdf.info()
```

```python
#값 치환
dropdf['Credit_History_Age'] = dropdf['Credit_History_Age'].astype(str).str.replace(' Years and ','.')
dropdf['Credit_History_Age'] = dropdf['Credit_History_Age'].astype(str).str.replace('Months','')
```

```python
dropdf['Payment_Behaviour'] = dropdf['Payment_Behaviour'].astype(str).str.replace('Low_spent_Small_value_payments','1')
dropdf['Payment_Behaviour'] = dropdf['Payment_Behaviour'].astype(str).str.replace('Low_spent_Medium_value_payments','2')
dropdf['Payment_Behaviour'] = dropdf['Payment_Behaviour'].astype(str).str.replace('Low_spent_Large_value_payments','3')
dropdf['Payment_Behaviour'] = dropdf['Payment_Behaviour'].astype(str).str.replace('High_spent_Small_value_payments','4')
dropdf['Payment_Behaviour'] = dropdf['Payment_Behaviour'].astype(str).str.replace('High_spent_Medium_value_payments','5')
dropdf['Payment_Behaviour'] = dropdf['Payment_Behaviour'].astype(str).str.replace('High_spent_Large_value_payments','6')
dropdf.head()
```

```python
#data type 변경
col_int2 = ['Age','Delay_from_due_date','Num_of_Delayed_Payment','Outstanding_Debt',
            'Total_EMI_per_month','Monthly_Balance','Payment_Behaviour','Credit_History_Age','Annual_Income']
for i in col_int2:
  dropdf[i] = dropdf[i].astype(float)
dropdf.dtypes
```

```python
#이상치 제거
dropdf['Age'] = dropdf['Age'].astype(int)
dropdf = dropdf[(dropdf['Age'] >= 0) & (dropdf['Age'] <= 100)]
dropdf['Age'].unique()
```

```python
dropdf['Credit_Score'].unique() #target
```

```python
dropdf['Payment_of_Min_Amount'].unique()
```

```python
#값 치환 : Good -> 3, Standard -> 2, Poor -> 1
dropdf['Credit_Score'] = dropdf['Credit_Score'].str.replace('Good', '3', n=-1)
dropdf['Credit_Score'] = dropdf['Credit_Score'].str.replace('Standard', '2', n=-1)
dropdf['Credit_Score'] = dropdf['Credit_Score'].str.replace('Poor', '1', n=-1)
dropdf['Credit_Score'] = dropdf[['Credit_Score']].apply(pd.to_numeric)

dropdf['Payment_of_Min_Amount'] = dropdf['Payment_of_Min_Amount'].str.replace('NM', '0')
dropdf['Payment_of_Min_Amount'] = dropdf['Payment_of_Min_Amount'].str.replace('Yes', '1')
dropdf['Payment_of_Min_Amount'] = dropdf['Payment_of_Min_Amount'].str.replace('No', '2')
dropdf['Payment_of_Min_Amount'] = dropdf[['Payment_of_Min_Amount']].apply(pd.to_numeric)
dropdf
```

```python
dropdf.dtypes
```

```python
dropdf = dropdf.drop_duplicates() #중복된 행 제거
dropdf.count()
```

# 모델 학습

```python
#학습시킬 데이터만 추출
select_df = cleandf[['Age', 'Annual_Income', 'Delay_from_due_date', 'Num_of_Delayed_Payment',
       'Outstanding_Debt', 'Credit_History_Age', 'Payment_of_Min_Amount',
       'Total_EMI_per_month', 'Payment_Behaviour', 'Monthly_Balance',
       'Credit_Score', 'Occupation_Accountant', 'Occupation_Architect',
       'Occupation_Developer', 'Occupation_Doctor', 'Occupation_Engineer',
       'Occupation_Entrepreneur', 'Occupation_Journalist', 'Occupation_Lawyer',
       'Occupation_Manager', 'Occupation_Mechanic', 'Occupation_Media_Manager',
       'Occupation_Musician', 'Occupation_Scientist', 'Occupation_Teacher',
       'Occupation_Writer']]
```

```python
#train, test 데이터 분할
from sklearn.model_selection import train_test_split
df_train_x = select_df.drop('Credit_Score',axis = 1)
df_train_y = select_df['Credit_Score']
x_train, x_test, y_train, y_test = train_test_split(df_train_x, df_train_y, test_size=0.20, random_state=42)
```

```python
#modeling
```

```python
#DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(random_state=1)
dtree = dtree.fit(x_train, y_train)

y_pred = dtree.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
accuracy = accuracy_score(y_test,y_pred)
print("confusion matrics=",cm)
print("  ")
print("accuracy=",accuracy)
```

```python
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(x_train, y_train)
kn_y_pred = kn.predict(x_test)
kn_cm = confusion_matrix(y_test,kn_y_pred)
kn_accuracy = accuracy_score(y_test,kn_y_pred)
print("confusion matrics=",kn_cm)
print("  ")
print("accuracy=",kn_accuracy)
```

```python
#RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=10)
rf.fit(x_train, y_train)
rf_y_pred = rf.predict(x_test)
rf_cm = confusion_matrix(y_test,rf_y_pred)
rf_accuracy = accuracy_score(y_test,rf_y_pred)
print("confusion matrics=",rf_cm)
print("  ")
print("accuracy=",rf_accuracy)
```

## 성능 평가

```python
#randomforest classfier 성능 평가
from sklearn.metrics import classification_report

target_names = ['Bad', 'Standard', 'Good']
print(classification_report(y_test,rf_y_pred, target_names=target_names))
```
